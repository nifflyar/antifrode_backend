```python
import pandas as pd
import numpy as np
from pathlib import Path

F1 = Path("Выгрузка 22.MQ.xlsx")
F2 = Path("Выгрузка 22.xlsx")
OUT = Path("merged_all.xlsx")
```

```python
def read_with_dynamic_header(path: Path, sheet_name=0, key_cols=("Номер билета", "Номер заказа", "Операция")):
    raw = pd.read_excel(path, sheet_name=sheet_name, header=None, dtype=str)
    header_row = 0

    # ищем строку, где встречаются ключевые названия столбцов
    for i in range(min(40, len(raw))):
        row = raw.iloc[i].astype(str).fillna("")
        hits = sum(any(k.lower() in str(cell).lower() for cell in row.values) for k in key_cols)
        exact = sum((row == k).sum() for k in key_cols)
        if hits >= 2 or exact >= 2:
            header_row = i
            break

    df = pd.read_excel(path, sheet_name=sheet_name, header=header_row)
    # удаляем Unnamed
    df = df.loc[:, ~df.columns.astype(str).str.startswith("Unnamed")]
    # удаляем пустые строки
    df = df.dropna(how="all").reset_index(drop=True)

    return df, header_row
```

```python
# 1) читаем оба файла
df1, h1 = read_with_dynamic_header(F1)
df2, h2 = read_with_dynamic_header(F2)

# 2) добавляем источник
df1["Источник"] = "MQ"
df2["Источник"] = "Full"

# 3) объединяем (align columns автоматически)
merged = pd.concat([df1, df2], ignore_index=True, sort=False)

# 4) сохраняем в один Excel
with pd.ExcelWriter(OUT, engine="openpyxl") as writer:
    merged.to_excel(writer, index=False, sheet_name="all")
    # короткая сводка на отдельный лист
    summary = pd.DataFrame({
        "Файл": ["Выгрузка 22.MQ.xlsx", "Выгрузка 22.xlsx", "ИТОГО"],
        "Строк": [len(df1), len(df2), len(merged)],
        "Header row (1-based)": [h1+1, h2+1, np.nan]
    })
    summary.to_excel(writer, index=False, sheet_name="summary")

print("OK ✅")
print("Saved:", OUT.resolve())
print("Rows:", len(merged), " | Columns:", merged.shape[1])
```

```python
import re
import hashlib
from pathlib import Path

INP = Path("merged_all.xlsx")
OUT = Path("clean_stage2.xlsx")

def normalize_text(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    s = re.sub(r"\s+", " ", s)
    return s

def parse_money(series: pd.Series) -> pd.Series:
    s = series.astype(str)
    s = s.str.replace("\u00a0", " ", regex=False).str.replace(" ", "", regex=False)
    s = s.str.replace(",", ".", regex=False)
    s = s.str.replace(r"[^0-9\.\-]", "", regex=True)
    return pd.to_numeric(s, errors="coerce")

def parse_dt(series: pd.Series) -> pd.Series:
    # dayfirst=True важно для наших дат
    return pd.to_datetime(series, errors="coerce", dayfirst=True)

def sha12(x: str) -> str:
    return hashlib.sha1(x.encode("utf-8")).hexdigest()[:12]

# 1) load
df = pd.read_excel(INP, sheet_name="all")

# 2) normalize columns + strings
df.columns = [normalize_text(c) for c in df.columns]
for c in df.columns:
    if df[c].dtype == object:
        df[c] = df[c].apply(normalize_text)

# 3) unify some names
if "Тип тарифа (Льгота)" in df.columns and "Тариф и льгота" not in df.columns:
    df["Тариф и льгота"] = df["Тип тарифа (Льгота)"]

if "Тип расчёта" in df.columns and "Тип расчета" not in df.columns:
    df["Тип расчета"] = df["Тип расчёта"]

# 4) money -> numeric
for col in ["Цена", "Разные сборы"]:
    if col in df.columns:
        df[col] = parse_money(df[col])

# 5) dates -> datetime
for col in ["Дата операции", "Дата/время отправки"]:
    if col in df.columns:
        df[col] = parse_dt(df[col])

# 6) derived fields
if "Станция отправления" in df.columns and "Станция назначения" in df.columns:
    df["Маршрут"] = df["Станция отправления"].fillna("") + " -> " + df["Станция назначения"].fillna("")
    df.loc[df["Маршрут"].str.replace("->", "").str.strip().eq(""), "Маршрут"] = np.nan

if "Дата/время отправки" in df.columns:
    df["Дата отправки"] = df["Дата/время отправки"].dt.date

if "Дата операции" in df.columns:
    df["Месяц операции"] = df["Дата операции"].dt.to_period("M").astype(str)
    df["Час операции"] = df["Дата операции"].dt.hour

if "Цена" in df.columns:
    fees = df["Разные сборы"] if "Разные сборы" in df.columns else 0
    df["Сумма_итого"] = df["Цена"].fillna(0) + pd.Series(fees).fillna(0)

# 7) passenger key + anonymized ID
# приоритет: документ -> ИИН -> (ФИО+телефон) -> Данные пассажира -> ФИО -> телефон
def build_key(row):
    doc = row.get("№ документа", np.nan)
    iin = row.get("ИИН", np.nan)
    fio = row.get("Ф.И.О. пассажира", np.nan)
    phone = row.get("Номер телефона", np.nan)
    pax = row.get("Данные пассажира", np.nan)

    if pd.notna(doc):   return f"DOC:{doc}"
    if pd.notna(iin):   return f"IIN:{iin}"
    if pd.notna(fio) and pd.notna(phone): return f"FIOPHONE:{fio}|{phone}"
    if pd.notna(pax):   return f"PAX:{pax}"
    if pd.notna(fio):   return f"FIO:{fio}"
    if pd.notna(phone): return f"PHONE:{phone}"
    return np.nan
```

```python
df["Пассажир_key"] = df.apply(build_key, axis=1)
df["Пассажир_ID"] = df["Пассажир_key"].apply(lambda x: sha12(x) if pd.notna(x) else np.nan)

# 8) basic QC
rows_before = len(df)
df = df.dropna(how="all").reset_index(drop=True)

print("Stage2 OK ✅")
print("Rows:", rows_before, "->", len(df), "| Cols:", df.shape[1])
print("Passengers (hashed):", df["Пассажир_ID"].nunique(dropna=True))

# 9) save
with pd.ExcelWriter(OUT, engine="openpyxl") as writer:
    df.to_excel(writer, index=False, sheet_name="all_clean")
    qc = pd.DataFrame({
        "metric": ["rows", "cols", "unique_passengers", "missing_passenger_id_share"],
        "value": [
            len(df),
            df.shape[1],
            df["Пассажир_ID"].nunique(dropna=True),
            float(df["Пассажир_ID"].isna().mean())
        ]
    })
    qc.to_excel(writer, index=False, sheet_name="qc")

print("Saved:", OUT.resolve())
```
