```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from pathlib import Path

# =========================
# CONFIG
# =========================
INPUT_FILE = "clean_stage2.xlsx"   # твой файл
OUTPUT_FILE = "final_results.xlsx"
SUSPICIOUS_FILE = "suspicious_passengers.xlsx"

# =========================
# 1. LOAD DATA

# =========================
df = pd.read_excel(INPUT_FILE)

# =========================
# 2. CLEAN DATA
# =========================

# оставляем только нужные операции
df = df[df["Операция"].isin(["Продажа", "Возврат"])]

# даты
df["Дата операции"] = pd.to_datetime(df["Дата операции"], errors="coerce")
df["Дата отправления"] = pd.to_datetime(df["Дата отправления"], errors="coerce")

# =========================
# 3. PASSENGER ID
# =========================
df["Пассажир_ID"] = (
    df["ФИО"].fillna("") +
    df["ИИН"].fillna("") +
    df["Документ"].fillna("")
).astype(str)

# =========================
# 4. FAKE FIO DETECTION
# =========================
def fake_fio_score(fio):
    if pd.isna(fio):
        return 1
    
    fio = str(fio).lower()
    
    # мало уникальных букв
    if len(set(fio)) <= 3:
        return 1
    
    # мусор
    if any(x in fio for x in ["ааа", "иии", "bbb", "test", "qwerty"]):
        return 1
    
    # короткое имя
    if len(fio) < 5:
        return 1
    
    return 0

df["fake_fio"] = df["ФИО"].apply(fake_fio_score)

# =========================
# 5. FEATURE ENGINEERING
# =========================

# ночь
df["hour"] = df["Дата операции"].dt.hour
df["is_night"] = df["hour"].between(0, 5).astype(int)

# дни активности
df["date_only"] = df["Дата операции"].dt.date

agg = df.groupby("Пассажир_ID").agg(
    total_tickets=("Номер билета", "count"),
    refund_cnt=("Операция", lambda x: (x == "Возврат").sum()),
    active_days=("date_only", "nunique"),
    night_tickets=("is_night", "sum"),
    fake_fio=("fake_fio", "max"),
    first_date=("Дата операции", "min"),
    last_date=("Дата операции", "max")
).reset_index()

# производные признаки
agg["refund_share"] = agg["refund_cnt"] / agg["total_tickets"]
agg["night_share"] = agg["night_tickets"] / agg["total_tickets"]

# =========================
# 6. SEAT BLOCKING DETECTION
# =========================

# возврат перед отправлением
df["hours_to_departure"] = (
    (df["Дата отправления"] - df["Дата операции"])
    .dt.total_seconds() / 3600
)

df["late_refund"] = (
    (df["Операция"] == "Возврат") &
    (df["hours_to_departure"] < 24)
).astype(int)

seat_block = df.groupby("Пассажир_ID").agg(
    late_refunds=("late_refund", "sum")
).reset_index()

agg = agg.merge(seat_block, on="Пассажир_ID", how="left")
agg["late_refunds"] = agg["late_refunds"].fillna(0)

# =========================
# 7. RULE BASED SCORE
# =========================

def rule_score(row):
    score = 0
    
    if row["total_tickets"] > 20:
        score += 20
        
    if row["refund_share"] > 0.5:
        score += 25
        
    if row["late_refunds"] > 5:
        score += 30
        
    if row["fake_fio"] == 1:
        score += 25
        
    if row["night_share"] > 0.5:
        score += 10
        
    return min(score, 100)

agg["rule_score"] = agg.apply(rule_score, axis=1)

# =========================
# 8. ML MODEL
# =========================

features = [
    "total_tickets",
    "refund_cnt",
    "refund_share",
    "night_share",
    "late_refunds",
    "fake_fio"
]

X = agg[features].fillna(0)

model = IsolationForest(
    n_estimators=300,
    contamination=0.03,
    random_state=42
)

model.fit(X)

agg["ml_raw"] = -model.decision_function(X)

# нормализация
agg["ml_score"] = (
    (agg["ml_raw"] - agg["ml_raw"].min()) /
    (agg["ml_raw"].max() - agg["ml_raw"].min() + 1e-9)
) * 100

# =========================
# 9. FINAL SCORE
# =========================

agg["final_score"] = (
    0.5 * agg["rule_score"] +
    0.5 * agg["ml_score"]
)

# =========================
# 10. RISK BAND
# =========================

def risk_band(score):
    if score > 80:
        return "CRITICAL"
    elif score > 60:
        return "HIGH"
    elif score > 40:
        return "MEDIUM"
    else:
        return "LOW"

agg["risk_band"] = agg["final_score"].apply(risk_band)

# =========================
# 11. SUSPICIOUS FILTER
# =========================

suspicious = agg[agg["risk_band"].isin(["HIGH", "CRITICAL"])]

# =========================
# 12. SAVE FILES
# =========================

with pd.ExcelWriter(OUTPUT_FILE) as writer:
    agg.to_excel(writer, sheet_name="All_Passengers", index=False)
    suspicious.to_excel(writer, sheet_name="Suspicious", index=False)

suspicious.to_excel(SUSPICIOUS_FILE, index=False)

# =========================
# DONE
# =========================

print("DONE ✅")
print("Total passengers:", len(agg))
print("Suspicious:", len(suspicious))
print("Saved:", OUTPUT_FILE, "|", SUSPICIOUS_FILE)
```
